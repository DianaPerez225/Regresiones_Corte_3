```python
import pandas as pd

datos = pd.read_excel("/content/drive/MyDrive/Estadística/Corte 3/Datos/datos_industrias.xlsx")
datos.head()
```

```python
# prompt: Necesito un análisis de regresión general simple, que la variables X sea "Empleados" y en Y sea "Tasa_Crecimiento" de nuestra base de datos "Datos" y genera una grafica sobre esa predición

import statsmodels.api as sm
import matplotlib.pyplot as plt

# Define dependent and independent variables
y = datos['Tasa_Crecimiento']
X = datos['Empleados']

# Add a constant to the independent variable for the intercept
X = sm.add_constant(X)

# Create and fit the OLS model
model = sm.OLS(y, X).fit()

# Print the summary of the regression
print(model.summary())

# Create the plot
plt.figure(figsize=(10, 6))
plt.scatter(datos['Empleados'], datos['Tasa_Crecimiento'], label='Datos Reales')
plt.plot(datos['Empleados'], model.predict(X), color='red', label='Línea de Regresión')
plt.xlabel('Empleados')
plt.ylabel('Tasa de Crecimiento')
plt.title('Análisis de Regresión: Tasa de Crecimiento vs Empleados')
plt.legend()
plt.grid(True)
plt.show()
```
